# TS2: Sistema Lineales e Invariantes en el Tiempo

## Introducción Teórica

En el análisis de señales y sistemas, los sistemas lineales e invariantes en el tiempo (LTI) ocupan un lugar central, ya que permiten describir, predecir y comprender el comportamiento de una amplia gama de procesos físicos y de procesamiento de señales. Estos sistemas cumplen dos propiedades fundamentales:

- Linealidad, que asegura que la respuesta a una combinación lineal de entradas es igual a la combinación lineal de las respuestas individuales.

- Invariancia en el tiempo, que implica que el comportamiento del sistema no cambia si la entrada se desplaza en el tiempo.

Una herramienta clave para describir sistemas LTI en el dominio discreto son las ecuaciones en diferencias lineales con coeficientes constantes, que establecen una relación entre la entrada x[n] y la salida y[n]. Estas ecuaciones permiten caracterizar tanto los sistemas FIR (Finite Impulse Response), en los que la salida depende de un número finito de muestras de la entrada, como los sistemas IIR (Infinite Impulse Response), donde además intervienen valores pasados de la propia salida.

La descripción más completa de un sistema LTI se logra a través de su respuesta al impulso, es decir, la salida que se obtiene cuando la entrada es una delta discreta δ[n]. Conociendo la respuesta al impulso h[n], la salida para cualquier señal causal puede determinarse mediante la operación de convolución discreta:

$$
y[n] = (x * h)[n] = \sum_{k=-\infty}^{\infty} x[k] \cdot h[n-k]
$$


En este marco, analizar la respuesta de un sistema frente a distintos tipos de señales (impulsos, escalones unitarios o senoidales) permite comprender sus características dinámicas y frecuenciales. Para señales periódicas, como las senoidales, resulta especialmente relevante la relación entre la entrada y la salida, ya que revela la respuesta en frecuencia del sistema.

Además, la caracterización de las señales de salida no se limita a su forma en el tiempo, sino que también se consideran magnitudes asociadas como la energía o la potencia, que ofrecen una medida cuantitativa del comportamiento del sistema en relación con la entrada aplicada.

En este trabajo se abordan distintos ejemplos de sistemas descritos por ecuaciones en diferencias, se determina su respuesta al impulso y se analizan las salidas correspondientes para señales causales, con el fin de aplicar y reforzar los conceptos fundamentales de los sistemas LTI.

## Análisis y Desarrollo

```python
import numpy as np
import matplotlib.pyplot as plt

def sen(f, n, A=1, fase=0, fs=2):
    N = np.arange(n)
    t = N/ fs
    x = A * np.sin(2 * np.pi * f * t + fase)
    return t, x

def impulso(N, n0=0):
    x = np.zeros(N)
    x[n0] = 1
    return x
```

### Primera Parte
Ecuación:

$$
y[n] = 0.03 \, x[n] + 0.05 \, x[n-1] + 0.03 \, x[n-2]
      + 1.5 \, y[n-1] - 0.5 \, y[n-2]
$$

Graficar la señal de salida para cada una de las señales de entrada que generó en el TS1. Considere que las mismas son causales.


```python
N = 200   
f = 2000         
fs = f*10                 
A = 1
fase = 0

t1, x1 = sen(f, N, A, fase, fs)

def sistema(x):
    y = np.zeros(len(x))

    for n in range(len(x)):
        xn   = x[n]
        if n-1 >= 0:
            xn1 = x[n-1]
        else:
            xn1 = 0

        if n-2 >= 0:
            xn2 = x[n-2]
        else:
            xn2 = 0

        if n-1 >= 0:
            yn1 = y[n-1]
        else:
            yn1 = 0

        if n-2 >= 0:
            yn2 = y[n-2]
        else:
            yn2 = 0
        
        y[n] = 0.03*xn + 0.05*xn1 + 0.03*xn2 + 1.5*yn1 + (-0.5)*yn2
    return y

y1=sistema(x1)


plt.figure(1)
plt.plot(t1*1e3,x1, label="x[n] entrada")
plt.plot(t1*1e3, y1, label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("1. Señal sinusoidal 2KHz")
plt.legend()
plt.grid(True)
```

Se puede observar el cumplimiento de una de las principales propiedades de un sistema LTI, una senoidal priduce una salida senoidal de la misma frecuencia, pero en este caso con menor amplitud y desfazada con respecto a la entrada. 

```python
#2)
N = 200   
f = 2000         
fs = f*10                 
A = 2
fase = np.pi/2

t2, x2 = sen(f, N, A, fase, fs)
y2=sistema(x2)

plt.figure(2)
plt.plot(t2*1e3,x2, label="x[n] entrada")
plt.plot(t2*1e3, y2, label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("2. Señal sinusoidal amplificada y desfazada")
plt.legend()
plt.grid(True)
```

Se obtiene un resultado similar a la señal anterior. 

```python
#3)
N=200
A=1
fs1=2000
B=1
fs2=fs1/2
Ts=1/(10*(fs1+fs2))
fs=1/Ts
t3=np.arange(0,0.01,Ts)

t3,s1 = sen(fs1,N,A,fase,fs) # portadora
t3,s2 = sen(fs2,N,B,fase,fs)   # moduladora
x3 =s1 * s2

y3=sistema(x3)

plt.figure(3)
plt.plot(t3*1e3,x3, label="x[n] entrada")
plt.plot(t3*1e3, y3, label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("3. Señal modulada en amplitud")
plt.legend()
plt.grid(True)
```

```python
#4)
N = 200   
f = 2000         
fs = f*10                 
A = 1
fase = 0

t4, x4 = sen(f, N, A, fase, fs)
Amax=A*0.75
y4=sistema(np.clip(x4,-Amax,Amax))

plt.figure(4)
plt.plot(t4*1e3,np.clip(x4,-Amax,Amax), label="x[n] entrada")
plt.plot(t4*1e3, y4, label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("4. Señal recortada 75%")
plt.legend()
plt.grid(True)
```

La señal de entrada no es una senoidal pura, al estar recortada se generan oscilaciónes extras además de la frecuencia fundamental. Se observa como el sistema suaviza la distorsión de la señal de entrada, y al igual que en los casos anteriores, esta tambien presenta una menor amplitud y un retardo de fase. 

```python
#5)
def cuad(f, n, A=1, fase=0, fs=2):
    N = np.arange(n)
    t = N / fs
    x = A * np.sign(np.sin(2 * np.pi * f * t + fase))
    return t, x

f=4000
fs=10*f
N=200
A=1
t5,x5=cuad(f,N,A,fase,fs)
y5=sistema(x5)

plt.figure(5)
plt.plot(t5*1e3,x5, label="x[n] entrada")
plt.plot(t5*1e3, y5, label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("5. Señal cuadrada 4KHz")
plt.legend()
plt.grid(True)
```

Como el resultado del paso de una señal cuadrada por nuestro sistema, se obtiene una señal distorsionada con curvas suaves y oscilaciones. Los armónicos que posee la señal de entrada son atenuados por el sistema.

```python
 #6)
def pulso_rect(n, A=1, fs=2, ancho=10e-3):
    N = np.arange(n)
    t = N / fs
    x = np.zeros(n)
    x[t < ancho] = A    
    return t, x


fs = 2000  
N  = 200
A  = 1

t6, x6 = pulso_rect(N, A, fs, ancho=10e-3)
y6 = sistema(x6)

plt.figure(6)
plt.plot(t6*1e3,x6, label="x[n] entrada")
plt.plot(t6*1e3, y6, label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("6. Pulso Rectangular 10ms")
plt.legend()
plt.grid(True)
```

Hallar la respuesta al impulso y usando la misma, repetir la generación de la señal de salida para alguna de las señales de entrada consideradas en el punto anterior.

La señal de entrada no es periódica como los casos anteriores, produce una respuesta distinta: en lugar de una salida senoidal atenuada y desfasada, aparece una respuesta que crece y se estabiliza en el tiempo.

```python
def respuesta_sist1(N,x):
   
    h = np.zeros(N)      # salida (respuesta al impulso)

    for n in range(N):
        h[n] = 0.03 * x[n]
        if n >= 1:
            h[n] += 0.05 * x[n-1] + 1.5 * h[n-1]
        if n >= 2:
            h[n] += 0.03 * x[n-2] - 0.5 * h[n-2]
    return h

N=200
x8=impulso(N,0)
f=2000
fs=10*f

h1=respuesta_sist1(N,x8)


S1=np.convolve(x1, h1)
S2=np.convolve(x2, h1)
S3=np.convolve(x3, h1)
S4=np.convolve(x4, h1)
S5=np.convolve(x5, h1)
S6=np.convolve(x5, h1)

plt.figure(8)
S1_rec = S1[:len(x1)]
plt.subplot(2,1,1)
plt.plot(t1*1e3, S1_rec, label="Salida por convolución")
plt.grid()
plt.legend()
plt.title("Señal 1")
plt.subplot(2,1,2)
plt.plot(t1*1e3,y1,color="orange", label="Salida ecuacion")
plt.grid()
plt.legend()

plt.figure(9)
S2_rec = S2[:len(x2)]
plt.subplot(2,1,1)
plt.plot(t2*1e3, S2_rec, label="Salida por convolución")
plt.grid()
plt.title("Señal 2")
plt.legend()
plt.subplot(2,1,2)
plt.plot(t2*1e3,y2,color="orange", label="Salida ecuacion")
plt.grid()
plt.legend()

plt.figure(10)
S3_rec = S3[:len(x3)]
plt.subplot(2,1,1)
plt.plot(t3*1e3, S3_rec, label="Salida por convolución")
plt.grid()
plt.title("Señal 3")
plt.legend()
plt.subplot(2,1,2)
plt.plot(t3*1e3,y3,color="orange", label="Salida ecuacion")
plt.grid()
plt.legend()

plt.figure(11)
S4_rec = S4[:len(x4)]
plt.subplot(2,1,1)
plt.plot(t4*1e3, S4_rec, label="Salida por convolución")
plt.grid()
plt.title("Señal 4")
plt.legend()
plt.subplot(2,1,2)
plt.plot(t4*1e3,y4,color="orange", label="Salida ecuacion")
plt.grid()
plt.legend()

plt.figure(12)
S5_rec = S5[:len(x5)]
plt.subplot(2,1,1)
plt.plot(t5*1e3, S5_rec, label="Salida por convolución")
plt.grid()
plt.title("Señal 5")
plt.legend()
plt.subplot(2,1,2)
plt.plot(t5*1e3,y5,color="orange", label="Salida ecuacion")
plt.grid()
plt.legend()

plt.figure(13)
S6_rec = S6[:len(x6)]
plt.subplot(2,1,1)
plt.plot(t6*1e3, S6_rec, label="Salida por convolución")
plt.grid()
plt.title("Señal 6")
plt.legend()
plt.subplot(2,1,2)
plt.plot(t6*1e3,y6,color="orange", label="Salida ecuacion")
plt.grid()
plt.legend()
```

Se verifica que al aplicar directamente la ecuación en diferencias o, de manera equivalente, al realizar la convolución de la respuesta al impulso con la señal de entrada, se obtiene el mismo resultado, con excepción del pulso rectangular.

### Segunda parte

Hallar la respuesta al impulso y la salida correspondiente a una señal de entrada senoidal en los sistemas definidos mediante las siguientes ecuaciones en diferencias:

#### Sistema 1
Ecuación:
$$
y[n] = x[n] + 3 \, x[n-10]
$$

```python
N=200
f=2000
fs=10*f
A=1
h2=impulso(N,0)+3*impulso(N,10)

t7,x7=sen(f,N,A,fase,fs)

Y=np.convolve(x7,h2)
Y_rec = Y[:len(x7)]

plt.figure(15)
plt.plot(t7*1e3, x7, label="x[n] entrada")
plt.plot(t7*1e3, Y_rec,  label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("Sistema 1")
plt.grid(True)
plt.legend()
```

Se obtiene una salida estable, con una amplitud mayor y desfasada con respecto a la señal de entrada. Únicamente depende de la entrada. 

#### Sistema 2
Ecuación:
$$
y[n] = x[n] + 3 \, y[n-10]
$$

```python
def resp_sist2(N,d):
    h = np.zeros(N) 
    for n in range(N):
       h[n] = d[n]
       if n >= 10:
          h[n] += 3* h[n-10]
    return h

N=200
d=impulso(N,0)
h3=resp_sist2(N,d)

Y2=np.convolve(x7, h3)
Y2_rec = Y2[:len(x7)]

plt.figure(17)
plt.plot(t7*1e3, x7, label="x[n] entrada")
plt.plot(t7*1e3, Y2_rec,  label="y[n] salida")
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("Sistema 2")
plt.grid(True)
plt.legend()


plt.show()
```

Cada muestra de salida depende de la entrada y también de una versión pasada de la propia salida. Cada 10 muestras el valor de salida se multiplca por tres, de esta manera crece de manera exponencial (se puede observar en el valor de la amplitud que la salida va tomando) debido al efecto de la retroalimentación. 

## Conclusión

A lo largo del trabajo se comprueba la importancia de la naturaleza de la señal de entrada. Cuando la entrada es periódica, se cumple la propiedad fundamental de los sistemas LTI: la salida es una señal senoidal de la misma frecuencia que la entrada. Lo que sí varía es la amplitud y el desfase, determinados respectivamente por el módulo y la fase de la respuesta en frecuencia, la cual proviene de la respuesta al impulso h[n]. 

En cambio, cuando la entrada es no periódica, la respuesta del sistema difiere de la observada con señales periódicas: ya no se obtiene una salida senoidal atenuada y desfasada, sino una señal que evoluciona de forma distinta, pudiendo crecer o estabilizarse en un valor constante dependiendo de las características del sistema y de la entrada.

Por otro lado se pudo verificar como la estructura del sistema puede determinar la estabilidad de la salida.
