# SINTESIS DE SE칌ALES

## Introducci칩n

La s칤ntesis de se침ales consiste en construir se침ales a partir de expresiones matem치ticas o combinaciones de formas de onda b치sicas. Permite generar, modificar y analizar se침ales en distintos dominios (tiempo, frecuencia, energ칤a).

### Se침ales elementales

Entre las formas de onda m치s utilizadas se encuentran:

- Sinusoidales (sin(洧랪洧노),cos(洧랪洧노)): base de la teor칤a de Fourier, fundamentales en modulaci칩n y transmisi칩n.

- Cuadradas y pulsos rectangulares: modelos ideales de se침ales digitales y ventanas temporales.

- Se침ales moduladas: productos de una portadora sinusoidal y una se침al moduladora (ej. modulaci칩n en amplitud).

Estas se침ales permiten estudiar propiedades como energ칤a, potencia, ortogonalidad y correlaci칩n.

### Operaciones sobre se침ales

- Escalamiento en amplitud: multiplica la se침al por un factor.

- Desfase temporal o de fase: altera el 치ngulo de una se침al peri칩dica.

- Recorte (clipping): limita la amplitud a un valor m치ximo.

- Modulaci칩n: genera nuevas frecuencias combinando dos se침ales (sumas y diferencias espectrales).

### Energ칤a y Potencia de se침ales

- Una se침al de energ칤a (no peri칩dica, duraci칩n finita) tiene energ칤a finita:
$ E= sum_{n=-inf}^{+inf} (|x[n]|^2)$

- Una se침al de potencia (peri칩dica, duraci칩n infinita) se caracteriza por su potencia media:
$ P= \lim_{N \to \infty} (sum_{n=-N}^{N}(|x[n]|^2))*(1/N)$

En los ejercicios, las sinusoides y se침ales peri칩dicas se analizan en t칠rminos de potencia, mientras que los pulsos finitos se caracterizan por su energ칤a.

### Ortogonalidad y Correlaci칩n

Dos se침ales son ortogonales si:

$ \int x(t)y(t)dt=0 $

La correlaci칩n extiende esta idea midiendo el grado de similitud entre dos se침ales bajo un desfase. La autocorrelaci칩n revela periodicidad, mientras que la correlaci칩n cruzada mide semejanza entre distintas se침ales.

### Identidades trigonom칠tricas y s칤ntesis

Las identidades de producto a suma permiten expresar productos de senos y cosenos como combinaciones de frecuencias suma y diferencia. En particular:

$2\sen(\alpha)\sen(\beta)=\cos(\alpha-\beta)-\cos(\alpha+\beta)$

Esto explica fen칩menos de batidos y de generaci칩n de frecuencias nuevas al multiplicar se침ales (como en la modulaci칩n AM). Verificar esta propiedad con se침ales muestreadas confirma la coherencia entre teor칤a y simulaci칩n.

## Desarrollo


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square

def energia(x):
    E = np.sum(np.abs(x)**2)
    return E

def potencia(x):
    N = len(x)
    return (1/N) * np.sum(np.abs(x)**2)

def productoInt(x, y, Ts):
    return np.sum(x * y) * Ts

# 1) Se침al sinusoidal
print("\n1) Se침al sinusoidal")
A=1
f1=2000
fs=10*f1
t=np.arange(0,0.01,1/fs)
phi=0

s1 = A*np.sin(2*np.pi*f1*t+phi)

plt.plot(t,s1)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.title("s1: Se침al Sinusoidal")
plt.show()

print("Ts=", 1/fs, "s")
print("N=", len(s1))
P=potencia(s1)
print("Potencia:", round(P,2),"W" )

#Autocorrelacion
Rs1s1=np.correlate(s1,s1, mode="full")
retardo = np.arange(-(len(s1)-1), len(s1))

plt.plot(retardo,Rs1s1, color="orange")
plt.title("Autocorrelacion")
plt.grid()
plt.show()


# 2) Se침al amplificada y desfasada
print("\n2) Se침al amplificada y desfasada")
A=3
f2=2000
fs=10*f2
t=np.arange(0,0.01,1/fs)
phi=np.pi/2

s2 = A*np.sin(2*np.pi*f2*t+phi)

plt.plot(t,s2, color="green")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.title("s2: Se침al Amplificada y Desfasada")
plt.show()

print("Ts=", 1/fs, "s")
print("N=", len(s2))
P=potencia(s2)
print("Potencia:", round(P,2),"W" )

# Ortogonalidad con s1
ref = np.sin(2*np.pi*f1*t+0)
O=productoInt(ref,s2,1/fs)
if(np.isclose(O,0)):
    print("Se침ales ortogonales")
else:
    print("Se침ales no ortogonales")  
    
#Correlacion
Rs1s2=np.correlate(s1,s2, mode="full")
retardo = np.arange(-(len(s1)-1), len(s2))

plt.plot(retardo,Rs1s2, color="orange")
plt.grid()
plt.title("Correlacion con S1")
plt.show()


# 3) Se침al modulada 
print("\n3) Se침al modulada")
A=1
f3=2000
phi=0
B=1
f4=f3/2
Ts=1/(10*(f3+f4))
t=np.arange(0,0.01,Ts)

s3a = A*np.sin(2*np.pi*f3*t+phi)   # portadora
s3b = B*np.sin(2*np.pi*f4*t+phi)   # moduladora
s3  = s3a * s3b

plt.subplot(3,1,1)
plt.plot(t,s3a, label="s1 original")
plt.legend(); plt.grid()

plt.subplot(3,1,2)
plt.plot(t,s3b, label="s3 moduladora", color="red")
plt.legend(); plt.grid()

plt.subplot(3,1,3)
plt.plot(t,s3, label="s3 modulada", color="orange")
plt.legend(); plt.grid()
plt.show()

print("Ts=", Ts, "s")
print("N=", len(s3))
P=potencia(s3)
print("Potencia:", round(P,2),"W" )

ref = np.sin(2*np.pi*f3*t+phi)
O=productoInt(ref,s3,Ts)
if(np.isclose(O,0)):
    print("Se침ales ortogonales")
else:
    print("Se침ales no ortogonales")   


#Correlacion
Rs1s3=np.correlate(s1,s3, mode="full")
retardo = np.arange(-(len(s1)-1), len(s3))

plt.plot(retardo,Rs1s3, color="orange")
plt.grid()
plt.title("Correlacion con S1")
plt.show()

# 4) Se침al recortada
print("\n4) Se침al recortada")
A=1
f5=2000
fs=10*f5
t=np.arange(0,0.001,1/fs)
phi=0

s4a = A*np.sin(2*np.pi*f5*t+phi)
Amax = 0.75*A
s4  = np.clip(s4a, -Amax, Amax)

plt.axhline(y=Amax, color="red", linestyle="--", label="A max")
plt.axhline(y=-Amax, color="red", linestyle="--")
plt.plot(t,s4)
plt.legend()
plt.title("s4: Se침al Recortada al 75%")
plt.grid()
plt.show()

print("Ts=", 1/fs,"s")
print("N=", len(s4))
P=potencia(s4)
print("Potencia:", round(P,2),"W" )

ref = np.sin(2*np.pi*f5*t+phi)
O=productoInt(ref,s4,1/fs)
if(np.isclose(O,0)):
    print("Se침ales ortogonales")
else:
    print("Se침ales no ortogonales")  
    
#Correlacion
Rs1s4=np.correlate(s1,s4, mode="full")
retardo = np.arange(-(len(s1)-1), len(s4))

plt.plot(retardo,Rs1s4, color="orange")
plt.grid()
plt.title("Correlacion con S1")
plt.show()

# 5) Se침al cuadrada
print("\n5) Se침al cuadrada")
f6 = 4000         
fs = 100*f6
t = np.arange(0,0.001,1/fs)

s5 = square(2*np.pi*f6*t)

plt.plot(t, s5)
plt.title("s5: Se침al Cuadrada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid()
plt.show()

print("Ts=", 1/fs,"s")
print("N=", len(s5))
P=potencia(s5)
print("Potencia:", round(P,2),"W" )

ref = np.sin(2*np.pi*2000*t+0)   
O=productoInt(ref,s5,1/fs)
if(np.isclose(O,0)):
    print("Se침ales ortogonales")
else:
    print("Se침ales no ortogonales")   

#Correlacion
Rs1s5=np.correlate(s1,s5, mode="full")
retardo = np.arange(-(len(s1)-1), len(s5))

plt.plot(retardo,Rs1s5, color="orange")
plt.title("Correlacion con S1")
plt.grid()
plt.show()

# 6) Pulso rectangular
print("\n6) Pulso rectangular")
N = 1000       
fs = 10000        # Hz (suficiente para una senoide de 2 kHz)       # n칰mero de muestras
t  = np.arange(N) / fs  
s6 = np.zeros(N)
for i in range(N):
    if (i >= 10) and (i <= 20):
        s6[i]=1


plt.plot(t, s6)
plt.xlabel("Tiempo [ms]")
plt.ylabel("Amplitud")
plt.title("s6: Pulso Rectangular")
plt.grid()
plt.show()

print("Ts= 1 ms")
print("N=", N)
E=energia(s6)
print("Energia:", round(E,2),"J")

ref = np.sin(2*np.pi*2000*t+0)   
O=productoInt(ref,s6,1/fs)
if(np.isclose(O,0)):
    print("Se침ales ortogonales")
else:
    print("Se침ales no ortogonales")  
    
#Correlacion
Rs1s6=np.correlate(s1,s6, mode="full")
retardo = np.arange(-(len(s1)-1), len(s6))

plt.plot(retardo,Rs1s6, color="orange")
plt.title("Correlacion con S1")
plt.grid()
plt.show()
```

### PROPIEDAD TRIGONOMETRICA

A) Demostraci칩n anal칤tica

$2\sen(\alpha)\sen(\beta)=\cos(\alpha-\beta)-\cos(\alpha+\beta)$

$\cos(\alpha-\beta)=\cos(\alpha)\cos(\beta) + \sen(\alpha)\sen(\beta)$
$\cos(\alpha+\beta)=\cos(\alpha)\cos(\beta) - \sen(\alpha)\sen(\beta)$

Se restan ambos cosenos y se obtiene $2\sen(\alpha)\sen(\beta)$

```python
#Propiedad Trigonometrica
#Demostraci칩n con se침ales

f=2000
fs=10*f
t=np.arange(0,0.01,1/fs )

s1=np.sin(2*np.pi*f*t)  #alfa
s2=np.sin(np.pi*f*t)  #beta
s3=np.cos(2*np.pi*f*t - np.pi*f*t)
s4=np.cos(2*np.pi*f*t + np.pi*f*t)

plt.subplot(2,1,1)
plt.plot(t,2*s1*s2, color="red", label="2sen(풤)sen()")
plt.title("Propiedad Trigonom칠trica")
plt.legend()
plt.grid()

plt.subplot(2,1,2)
plt.plot(t,(s3-s4),color="green", label="cos(풤-)-cos(풤+)")
plt.legend()
plt.grid()
plt.show()
```

Se puede observar que ambas se침ales son identicas, cumpliendo asi la propiedad.
